```python
import pandas as pd
import numpy as np
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic[['sex', 'age', 'class', 'alive']]
```

`sort_values` 메서드를 사용하여 타이타닉호 승객에 대해 성별(sex) 인원수, 나이별(age) 인원수, 선실별(class) 인원수, 사망/생존(alive) 인원수를 구하라.

```python
# 원하는 항목의 리스트를 만든 다음 for문을 돌려서 타이타닉 데이터프레임에서 각각의 정보를 불러와보자
a = ['sex', 'age', 'class', 'alive']
for i in a :
    print(titanic[i].value_counts())
```

3. 타이타닉호 승객중 1등실 선실의 여성 승객의 평균 나이를 구하라.

```python
# 1 타이타닉호 승객의 평균 나이를 구하라.
round(titanic['age'].mean(),2)
```

```python
# 2 타이타닉호 승객중 여성 승객의 평균 나이를 구하라.
round(titanic[titanic['sex']=='female']['age'].mean(),2)
```

```python
# 3 타이타닉호 승객중 1등실 선실의 여성 승객의 평균 나이를 구하라.
round(titanic[(titanic['class']=='First') & (titanic['sex']=='female')]['age'].mean(),2)
```

타이타닉호의 승객 중 나이 20살을 기준으로 성인(adult)과 미성년자(child)를 구별하는 라벨 열을 만들어 보자.

```python
# lambda 사용
titanic["adult/child"] = titanic.apply(lambda i : 'adult' if i.age >= 20 else "child", axis=1)
titanic.head(10)
```

```python
# 리스트컴프리헨션 사용
titanic["adult/child_2"] = ['adult' if x >=20 else 'child'  for x in titanic['age']]
titanic.head(10)
```

타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 `category1` 열을 만들어라.   
`category1` 카테고리는 다음과 같이 정의된다.

1. 20살이 넘으면 성별을 그대로 사용한다.
2. 20살 미만이면 성별에 관계없이 "child"라고 한다.

```python
titanic["category1"] = titanic.apply(lambda i : i.sex if i.age >= 20 else "child", axis=1)
titanic.tail(10)
```

타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고쳐라.


```python
int(titanic['age'].mean())
```

```python
titanic['age'] = titanic['age'].fillna(round(titanic['age'].mean()))
titanic.tail(10)
```

타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 `category2` 열을 만들어라.   
`category2` 카테고리는 다음과 같이 정의된다.

1. 성별을 나타내는 문자열 `male` 또는 `female`로 시작한다.
2. 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열이 온다.
3. 예를 들어 27살 남성은 `male27` 값이 된다.

```python
titanic["category2"] = titanic.apply(lambda i : (i.sex + str(i.age)) if i.age >= 20 else "child", axis=1)
titanic.tail(10)
```

```python
# 자료형을 맞춘 다음 + 로 붙여서 출력
titanic['category2'] = titanic.sex + titanic.age.astype(str)
titanic
```

```python
titanic["category2"] = titanic.apply(lambda x: x.sex + str(int(x.age)), axis=1)
titanic
```

타이타닉호 승객을 '미성년자', '청년', '중년', '장년', '노년' 나이 그룹으로 나눈다.

bins = [1, 20, 40, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]

그리고 각 나이 그룹의 승객 비율을 구한다. 비율의 전체 합은 1이 되어야 한다.

```python
bins = [1, 20, 40, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]

titanic['age2'] = pd.cut(titanic.age, bins, labels=labels)
titanic.head(10)
```

```python
h=pd.qcut(titanic.age, 5, labels= ["미성년자", "청년", "중년", "장년", "노년"])
pd.value_counts(h)
```
